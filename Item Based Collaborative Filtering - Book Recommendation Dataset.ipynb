```python
df["User-ID"] = df["User-ID"].astype(object)
```

```python
comment_counts=pd.DataFrame(df["ISBN"].value_counts())
```

```python
rare_books = comment_counts[comment_counts["ISBN"]>100].index
```

```python
comment_books = df[df["ISBN"].isin(rare_books)]
```
